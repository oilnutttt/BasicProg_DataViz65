## 2 รวมคะแนนและตัดเกรดแบบ อิงเกณฑ์ <50 ได้ F ... >= 80 ได้ A และ save ตาราง ชื่อ คะแนนรวมและเกรดที่ได้ของนักศึกษาแต่ละคน
*คะแนนรวม 60 คะแนน (ดึงข้อมูลและรวมข้อมูล 40 ตัดเกรด 20)* 

```python
total_score = advisor.merge(DAYY,left_on='นักศึกษา', right_on='นักศึกษา')
total_score
```

```python
total_score['คะแนนรวม'] = (total_score['คะแนนรวม 60%']+total_score['คะแนนรวม 40%'])
```

```python
total_score
```

```python
import numpy as np
```

```python
total_score['Grade'] = np.where((total_score.คะแนนรวม < 50 ), 
                  'F', np.where((total_score.คะแนนรวม >= 50) & (total_score.คะแนนรวม <55),
                  'D', np.where((total_score.คะแนนรวม >= 55) & (total_score.คะแนนรวม <60), 
                  'D+', np.where((total_score.คะแนนรวม >= 60) & (total_score.คะแนนรวม <65),  
                  'C', np.where((total_score.คะแนนรวม >= 65) & (total_score.คะแนนรวม <70), 
                  'C+', np.where((total_score.คะแนนรวม >= 70) & (total_score.คะแนนรวม <75), 
                  'B', np.where((total_score.คะแนนรวม >= 75) & (total_score.คะแนนรวม <80),
                  'B+', np.where((total_score.คะแนนรวม >= 80) & (total_score.คะแนนรวม <=100), 
                  'A', 'No Marks'))))))))
total_score
```
