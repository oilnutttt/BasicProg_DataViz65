## HW2
1. เปรียบเทียบราคาเฉลี่ยของ Detached house, Condo, Twin house และ Townhome
2. พื้นที่ (area_total_min_wa และ area_usable_min) และ จำนวนห้องนอนของทรัพย์สินที่แพงที่สุดแต่ละชนิด
3. จำนวนที่จอดรถและห้องน้ำของของทรัพย์สินที่ถูกที่สุดแต่ละชนิด

```python
dmissing = unit.dropna(subset=['price_min','area_total_min_wa','area_usable_min','count_room_bed','count_parking','count_room_bath']) # ลบ missing
```

### 1.เปรียบเทียบราคาเฉลี่ยของ Detached house, Condo, Twin house และ Townhome

```python
unit.groupby(['propertytype_name_en'])['price_min'].mean()[['Detached House','Condo','Twin house','Townhome']] #  groupby ชนิดของบ้าน เลือก column price_min ใช้.mean()หาค่าเลี่ย
```

### 2.พื้นที่ (area_total_min_wa และ area_usable_min) และ จำนวนห้องนอนของทรัพย์สินที่แพงที่สุดแต่ละชนิด

```python
dmissing.groupby(['propertytype_name_en']).max()[['area_total_min_wa','area_usable_min','count_room_bed','price_min']]
 #Groupby ชนิดของบ้านแต่ละชนิด ใช้.max() หาค่าแพงที่สุด ใช้list เลือกเฉพาะ column พื้นที่และจำนวนห้องนอน
```

###3.จำนวนที่จอดรถและห้องน้ำของของทรัพย์สินที่ถูกที่สุดแต่ละชนิด

```python
dmissing.groupby(['propertytype_name_en']).min()[['count_parking','count_room_bath','price_min']]
```
